```python
"""
This code was used to cheat on the NYTimes Spelling Bee on 2021-07-25. I got to "genius" and then
was stuck, so I was curious to find the words I had missed.

Using https://norvig.com/ngrams/word.list as the potential valid words, it returns a list of matches
against Spelling Bee's rules (> 3 characters, composed of input letters, must have the middle letter).

This code returns 59 matching words of which the puzzle recognized 26.

To learn more:
- https://www.nytimes.com/puzzles/spelling-bee
- https://slate.com/human-interest/2020/02/spelling-bee-new-york-times-praise.html
"""

letters = 'belmnow'
required = 'm'

with open('data/words.txt') as f:
    words = [word for line in f.readlines()
             if len(word := line.strip()) > 3
             if required in word
             if all(letter in letters for letter in word)]

for word in sorted(words):
    if all(letter in word for letter in letters):
        print('* ', end='')
        
    print(word)
```
